## Part 1

For simplicity for this dataset, I am going to treat all of the features as continuous values instead of mixing and matching 
discrete and continuous. Therefore, to calculate the conditional probability distribution for these features, we can assume they have a normal distribution and we can use the Gaussian Distribution equation.


![](normal_dist_eq.png)

To calculate the normal distribution I used excel to calculate all of the means and standard deviations for each feature with each respective class. These calculations yielded 24 means and 24 standard deviations.

![](mean_std.png)

Since I am treating the data as continuous, the last step I have to do is plug the x value for each feature into the distribution equation along with the respective mean and standard deviation for that feature and class. This step will be done for part 2.

## Part 2


```python
import math

apartment_std = [3.347762921,0.524890659,3.016793588,0.651875317,0.646813224,1.245399698,0.903507903,13.59321559]
apartment_mean = [7.332742857,1.285714286,6.103857143,1.505,1.214285714,6.857142857,3.428571429,38.71428571]
apartment_class_prob = 7 / 20

condo_std = [4.209474117,0.552770798,2.323053283,0.842982733,0.471404521,1.462494065,0.745355992,12.73664878]
condo_mean = [7.4159,1.333333333,6.024666667,1.553333333,1.333333333,6.833333333,3.333333333,39.66666667]
condo_class_prob = 6 / 20


house_std = [0.527829731,0.174963553,2.082144609,0.197129192,0.775912892,0.638876565,0.534522484,10.69999046]
house_mean = [5.760742857,1.071428571,6.6309,1.391714286,1.071428571,6.142857143,3,35.28571429]
house_class_prob = 7 / 20



test_data = [[6.0931,1.5,6.7265,1.652,1,6,3,44,'Apartment'],
             [8.3607,1.5,9.15,1.777,2,8,4,48,'House'],
             [8.14,1,8,1.504,2,7,3,3,'House'],
             [9.1416,1.5,7.3262,1.831,1.5,8,4,31,'Apartment'],
             [12,1.5,5,1.2,2,6,3,30,'Apartment']]

cond_probs_apart = []
cond_probs_condo = []
cond_probs_house = []

class_predictions = []

for row in range(len(test_data)):
    
    t_cond_probs_apart = []
    t_cond_probs_condo = []
    t_cond_probs_house = []

    for feature in range(8):
        cond_prob = (1 / (apartment_std[feature] * math.sqrt(2 * math.pi)) * math.exp( ( -(test_data[row][feature] - apartment_mean[feature]) ** 2 / (2 * (apartment_std[feature]) ** 2) )) )
        t_cond_probs_apart.append(cond_prob)    

    for feature in range(8):
        cond_prob = (1 / (condo_std[feature] * math.sqrt(2 * math.pi)) * math.exp( ( -(test_data[row][feature] - condo_mean[feature]) ** 2 / (2 * (condo_mean[feature]) ** 2) )) )
        t_cond_probs_condo.append(cond_prob) 
        
    for feature in range(8):
        cond_prob = (1 / (house_std[feature] * math.sqrt(2 * math.pi)) * math.exp( ( -(test_data[row][feature] - house_mean[feature]) ** 2 / (2 * (house_mean[feature]) ** 2) )) )
        t_cond_probs_house.append(cond_prob)
        
    cond_probs_apart.append(t_cond_probs_apart)
    cond_probs_condo.append(t_cond_probs_condo)
    cond_probs_house.append(t_cond_probs_house)
    
    apart_class_pred = 1
    condo_class_pred = 1
    house_class_pred = 1
    
    for i in range(8):
        apart_class_pred = apart_class_pred * cond_probs_apart[row][i]
        condo_class_pred = condo_class_pred * cond_probs_condo[row][i]
        house_class_pred = house_class_pred * cond_probs_house[row][i]
        
    apart_class_pred = apart_class_pred * apartment_class_prob
    condo_class_pred = condo_class_pred * condo_class_prob
    house_class_pred = house_class_pred * house_class_prob
        
    class_preds_by_row = [apart_class_pred, condo_class_pred, house_class_pred]
    class_predictions.append(class_preds_by_row)
    
   
print('======================================================\n             CONDITIONAL PROBABILITY TABLE\n======================================================\n')
for i in range(5):
    print(f'Cond. Prob. for input: {test_data[i]}') 
    print('Apartment') 
    print(cond_probs_apart[i])

    print('Condo') 
    print(cond_probs_condo[i])
    print('House') 
    print(cond_probs_house[i])
    print('\n')
    
print('======================================================\n             CLASS PREDICTIONS TABLE\n======================================================\n')
for i in range(5):
    print(f'Class Pred. for input: {test_data[i]}') 
    print(f'{class_predictions[i]}\n')
```

```python
#Importing the libraries  

import numpy as np 
import matplotlib.pyplot as mtp  
import pandas as pd  

# Importing the data
data= pd.read_csv('train.csv')  
x_train = data.iloc[:, 1:9].values  
y_train = data.iloc[:, 9].values  


#taking a look at the data

data.head()

#splitting the data into training and testing sets

# from sklearn.model_selection import train_test_split  
# x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 5)

# Importing the data
data1= pd.read_csv('test.csv')  
x_test = data1.iloc[:, 1:9].values  
y_test = data1.iloc[:, 9].values  


# Feature Scaling  

from sklearn.preprocessing import StandardScaler  
sc = StandardScaler()  
x_train = sc.fit_transform(x_train)  
x_test = sc.transform(x_test) 

# Fitting Naive Bayes to the Training set  

from sklearn.naive_bayes import GaussianNB  
classifier = GaussianNB()  
classifier.fit(x_train, y_train)  

# Predicting the Test set results  

y_pred = classifier.predict(x_test)  

print(y_pred)

from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_pred) 
```
